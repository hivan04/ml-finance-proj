# Return Objects

Overnight Return:
$$
r_{ON,t}=(\frac{\text{Open}_t}{\text{Close}_{t-1}})-1
$$

Intraday Return:
$$
r_{ID,t}=(\frac{\text{Open}_t}{\text{Close}_{t}})-1
$$

Close-to-Close Return:
$$
(1+r_{ON,t})\times(1+r_{ID,t})-1=(\frac{Close_t}{Close_{t-1}})-1
$$
